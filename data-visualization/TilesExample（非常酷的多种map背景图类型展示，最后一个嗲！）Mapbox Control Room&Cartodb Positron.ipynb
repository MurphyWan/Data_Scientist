### 06）Folium支持来自Leaflet-DVF的多边形标注(Maker)的设置:

```python



lon, lat = 31.16885, 121.39349

tiles = "http://{s}.tiles.wmflabs.org/bw-mapnik/{z}/{x}/{y}.png" #, {maxZoom= 16, minZoom= 5}
zoom_start = 14
max_zoom = 18
attrwbt = '&copy; <a href="http://www.5156bao.com" target="_blank">旺保通</a>'

map_6 = folium.Map(location = [lon, lat],
                   tiles = tiles,
                   zoom_start = zoom_start,
                   attr = attrwbt)

folium.RegularPolygonMarker(
    [31.16885, 121.39349],
    popup='YiBao Network',
    fill_color='#F79709',
    number_of_sides=3,
    radius=10
    ).add_to(map_6)
folium.RegularPolygonMarker(
    [31.16885, 121.3800],
    popup='YiBao Network2',
    fill_color='#F79709',
    number_of_sides=4,
    radius=10
    ).add_to(map_6)
folium.RegularPolygonMarker(
    [31.16885, 121.3900],
    popup='YiBao Network3',
    fill_color='#F79709',
    number_of_sides=6,
    radius=10
    ).add_to(map_6)
folium.RegularPolygonMarker(
    [31.16885, 121.3700],
    popup='YiBao Network4',
    fill_color='#F79709',
    number_of_sides=8,
    radius=10
    ).add_to(map_6)
map_6
```

### 07）可以通过folium.Vega(json.load('xxx.json'))可以将Vincent可视化图表传递给任一一个标注（maker）
####  notes：需要导入json库
####  Folium enables passing vincent visualizations to any marker type, with the visualization as the popover:

```python
#导入json库
import json

#以下定义基图部分

lon, lat = 31.16885, 121.39349

tiles = "http://{s}.tiles.wmflabs.org/bw-mapnik/{z}/{x}/{y}.png" #, {maxZoom= 16, minZoom= 5}
zoom_start = 17
max_zoom = 18
attrwbt = '&copy; <a href="http://www.5156bao.com" target="_blank">旺保通</a>'

map_7 = folium.Map(location = [lon, lat],
                   tiles = tiles,
                   zoom_start = zoom_start,
                   attr = attrwbt)
#—————————————————————关键代码部分————————————————————

folium.RegularPolygonMarker(
                            [lon, lat],
                            fill_color='#F79709',
                            radius=12,
                            popup=folium.Popup(max_width=450).add_child(
                                                                        folium.Vega(json.load(open('vis1.json')), #folim.Vega
                                                                                    width= 450 , 
                                                                                    height= 250))
                                                                        ).add_to(map_7)
                            #popup(folium.Popup)-->map类型(add_child)-->加到具体的基图上(此处是map_7)。

#———————————————————————————————————————————————


map_7.save('03-Vincent-Vega.html')
map_7
```

### 08) 在弹出窗口（popup）中显示内容（如：文字）

```python
lon, lat = 31.16885, 121.39349

tiles = "http://{s}.tiles.wmflabs.org/bw-mapnik/{z}/{x}/{y}.png" #, {maxZoom= 16, minZoom= 5}
zoom_start = 17
max_zoom = 18
attrwbt = '&copy; <a href="http://www.5156bao.com" target="_blank">旺保通</a>'

map_8 = folium.Map([lon, lat], 
                   tiles= tiles,
                   zoom_start= zoom_start,
                   attr= attrwbt  
                  )
#—————————————————————关键代码部分————————————————————

text = folium.Html('<p>Hi, Here is the location of WBT</p>\
                   <p>We specialize in customizing the insurance products of traffic scene</p>',
                   script=True)                                                                    #定义文本的内容

popup = folium.Popup(text, max_width=2650)             #定义popup

folium.RegularPolygonMarker(                           #定义多边形maker
                            location=[lon, lat],
                            fill_color ="#F79709",
                            number_of_sides = 8,
                            popup=popup,
                            ).add_to(map_8)
#—————————————————————关键代码部分————————————————————

map_8
```
