# 1/4）物流行业
## 物流行业体量之大超乎你我的想象,远超同期的房产、医疗、教育的体量。
- 2016年全国社会物流总额（物品）229.7万亿元。社会物流总费用11.1万亿元
- 其中公路物流占75%达8万亿，超同期的医疗、教育等行业的体量


```python
import pyecharts
from pyecharts import Pie


attr = ["工业品物流总额", "进口货物物流总额", "农产品物流总额", "再生资源物流总额", "单位与居民物流总额"]
v1 = [214, 10.5, 3.6, 0.9, 0.7]
pie_Lindustry = Pie("2016年全国社会物流总额229.7万亿元图示例 单位:亿和%", title_pos='center', width=900)
pie_Lindustry .add("社会物流总额A", attr, v1, 
                   center=[50, 55], #居x、y中部，%
                   label_color= ['#FF9900','#FFFF00','#138FCB','#0B40B7','#A5A5A5'],#FFFF00
                   is_random=False, 
                   radius=[35, 66], #居内圈、外圈半径大小
#                    rosetype='area', #确定饼图类型，不写就是正常比例；写radius就是根据半径大小；area也是饼图平分面积，但是半径根据数值来。
#                    rosetype='area',
                   legend_top ='bottom', #legend位置，在底部
                   is_label_show = True)
#pie_Lindustry .add("社会物流总额B", attr, v1, center=[75, 50], is_random=True, radius=[30, 75], rosetype='area',is_label_show=True)
#pie_Lindustry .show_config() 
pie_Lindustry.render('pie_Lindustry.html')
pie_Lindustry
```

# 物流行业—物流费用占比GDP的比重很高
- 下图显示，我国物流费用占比GDP有逐年下降的趋势，但是与美国相比仍旧有近一倍左右的差距。于2016年达到14.8%。国际上通常用物流费用在GDP中的占比来衡量一个国家物流发展水平。

```python
from pyecharts import Line

attr= ['2011年', '2012年','2013年','2014年','2015年','2016年']
v1 = [17.4,17.5,16.9,16.6,16.0 ,14.8] #不能直接使用百分数，数值型
v2 = [8.3,8.5,8.4,8.4,8.3 ,8.0]
line = Line('2011-2016年中美物流总费用占GDP比重 (单位:%)',
            title_pos='center')
line.add('中国占比',attr, v1,
         mark_point=["average"],
         legend_top ='bottom', 
         is_fill = False,          #is_fill = False , the chart looks like a "line chart"
         label_color= ['#FF9900'],
         is_label_show = True,
         line_width=3)

line.add('美国占比',attr, v2,
         mark_point=["average"],
         lenged_top = 'bottom' ,
         is_fill =False,            #is_fill = True , the chart looks like a "square"
         label_color= ['#138FCB'], 
         legend_top ='bottom',     #如果有两条Line，bar等，那么要放置图例位置，这时要将legend_top参数放置在最后一个图案中
         is_label_show = True,
         line_width=3)

line.render('line.html')
line
```
